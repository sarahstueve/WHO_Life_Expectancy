Countries fixed by using a dictionary (matches are too different and don't follow same patterns)

```python
overlap_dict = {'Czech Republic':"Czechia", 
               'United Kingdom': 'United Kingdom of Great Britain and Northern Ireland', 
               'United States': 'United States of America',
               'Macedonia (the former Yugoslav Republic of)':'The former Yugoslav republic of Macedonia'}
```

Pattern matching and building new list of countries (so data matches):

```python
overlap = 0
not_in_expect = []
new_countries = []
for country in codes['english_name'].values:
    country2 = ""
    if "," in country:
        country = " (".join(country.split(", ")) + ")"
        country_split = country.replace("(", "").replace(")", "").split()
        country2 = country_split[1:].copy()
        country2.append(country_split[0])
        country2 = " ".join(country2).strip("the ")
    if country in life_expectancy.Country.values:
        overlap += 1
        new_countries.append(country)
    elif country2 and country2 in life_expectancy.Country.values:
        overlap += 1
        new_countries.append(country2)
    elif country in overlap_dict:
        overlap += 1
        new_countries.append(country)
    else:
        new_countries.append(country)
        not_in_expect.append(country)
print("Proportion of countries matching:",overlap)
print()
print("Countries not in life expectancy data:", not_in_expect)
```

```python
name_to_code = {}
code_to_name = {}
for row in codes.index:
    name_to_code[codes.loc[row, "english_name"]] = codes.loc[row, "alpha-2_code"]
    code_to_name[codes.loc[row, "alpha-2_code"]] = codes.loc[row, "english_name"]
print(name_to_code)   
```

```python
iso_codes = []
for row in life_expectancy.index:
    country_name = life_expectancy.loc[row, "Country"]
    if country_name in name_to_code:
        iso_codes.append(name_to_code[country_name])
    else:
        iso_codes.append("")
```

Extract a dataframe for each year and save to new csv and json files:

```python
for year in life_expectancy.Year.unique():
    year_df = life_expectancy[life_expectancy.Year == year].copy()
    year_df.reset_index(drop = True, inplace = True)
    year_df.to_csv("data/life_expect_by_year_csv/life_expect_" + str(year) + ".csv", sep = ";", index = False)
    year_df.to_json("data/life_expect_by_year_json/life_expect_" + str(year) + ".json", orient = "index")
```

Read in geojson data and add properties for variables to visualize.

```python
import requests
url = 'https://raw.githubusercontent.com/datasets/geo-countries/master/data/countries.geojson'
f = requests.get(url)
# The .json() method automatically parses the response into JSON.
geo_data = f.json()
```

```python
geo_data.keys()
```

```python
geo_data['features'][1]
```

```python
import os, json
path = 'data/life_expect_by_year_json'
year_dicts = {}
for file in os.listdir(path):
    if file[-5:] == '.json':
        year_json_orig = json.load(open(path + '/' + file))
        year = year_json_orig['0']['Year']
        for key in year_json_orig:
            if year not in year_dicts:
                year_dicts[year] = {}
            if year_json_orig[key]['iso_alpha-2'] != '' and year_json_orig[key]['iso_alpha-2'] is not None:
                year_dicts[year].update({year_json_orig[key]['iso_alpha-2']: {var.strip().lower().replace(' ', '_'): year_json_orig[key][var]
                                                                         for var in ['Status', 
                                                                                     'Life expectancy ', 
                                                                                     'Adult Mortality', 
                                                                                     'infant deaths',
                                                                                     'Alcohol',
                                                                                     'Population',
                                                                                     'Schooling']}})
```

```python
null_props_dict = {key: None for key in year_dicts[2015]['AZ'].keys()}
null_props_dict
```

```python
import copy
for year in year_dicts:
    geo_data_copy = copy.deepcopy(geo_data)
    for feature in geo_data_copy['features']:
        iso_a2 = feature['properties']['ISO_A2']
        print(iso_a2)
        if iso_a2 in year_dicts[year]:
            add_props = year_dicts[year][iso_a2]
            print(add_props)
            feature['properties'].update(add_props)
        else:
            feature['properties'].update(null_props_dict.copy())
    with open('data/geo_json/life_expect_country_' + str(year) + '.json', 'w') as fp:
        json.dump(geo_data_copy, fp)
```
